```python

#LAB 6 Otsu thresholding experiment
import cv2
import numpy as np
import matplotlib.pyplot  as plt
img=cv2.imread("/content/posing-women-in-photos.webp",0)
if img is None:
  raise ValueError("Image not found")

threshold_value,binary_img=cv2.threshold(
    img,
    100,
    255,
    cv2.THRESH_BINARY +cv2.THRESH_OTSU


)
print("Optimal Threshold Value:",threshold_value)
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img,cmap='gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Histogram")
plt.hist(img.ravel(),256)
plt.xlabel("Pixel Value")

plt.hist(img.ravel(),256)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(1,3,3)
plt.title("Otsu Binarized Image")
plt.imshow(binary_img,cmap='gray')
plt.axis("off")

plt.show()

```
