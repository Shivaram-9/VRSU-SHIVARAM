```python
#LAB 2 CLAHE ENHANCED
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("/content/posing-women-in-photos.webp",0)
clahe=cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_img=clahe.apply(img)
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img,cmap='Set3')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("CLAHE ENHANCED")
plt.imshow(clahe_img, cmap='cool')
plt.axis("off")
```
