```python
#Point processing(Lab1)
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("/content/posing-women-in-photos.webp",0)#reading the image
he_img = cv2.equalizeHist(img)
plt.figure(figsize=(12,5))#size of histogram i.e. x and y length
plt.subplot(1,3,1)#dimensions of the original image
plt.title("Original")
plt.imshow(img,cmap = 'gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Clarified image")
plt.imshow(he_img,cmap ='gray')
plt.axis("off")#axis point numbers won't be visible

plt.subplot(1,3,3)
plt.title("Graph")#in the graph the y-axis indicates the white color pixels and x-axis indicates the gray color
plt.hist(he_img.ravel(),256)
plt.show()

```
